```python
import dg_geometry as DG503
import sympy as sp
import numpy as np
```

```python
x1, x2, x3, f = sp.symbols('x1 x2 x3 f', real=True)
coords = [x1, x2, x3]
f = sp.Function('f')(x1, x2, x3)
GK = f**2*sp.Matrix([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
display(GK)
f1 = 1/x3
GK_f1 = GK.subs(sp.Function('f')(x1, x2, x3), f1)
display(GK_f1)
K_f1 = DG503.sectional_curvature(GK_f1, sp.Matrix([[x1], [x2], [x3]]), sp.Matrix([[x1], [x2], [x3]]), sp.Matrix([[-x2], [x1], [x3]]), coords)
display(sp.simplify(K_f1))
```

```python
x1, x2, x3, f = sp.symbols('x1 x2 x3 f', real=True)
coords = [x1, x2, x3]
f = sp.Function('f')(x1, x2, x3)
GK = sp.Matrix([[1, 0, 0], [0, 1, 0], [0, 0, f**2]])
display(GK)

f1 = 1/x3
GK_f1 = GK.subs(sp.Function('f')(x1, x2, x3), f1)
display(GK_f1)

S = DG503.scalar_curvature(GK_f1,sp.Matrix([x1,x2,x3]), coords)
display(sp.simplify(S))

K_12 = DG503.sectional_curvature(GK_f1, sp.Matrix([[x1], [x2], [x3]]), sp.Matrix([1, 0, 0]), sp.Matrix([0, 1, 0]), coords)
display(sp.simplify(K_12))

K_13 = DG503.sectional_curvature(GK_f1, sp.Matrix([[x1], [x2], [x3]]), sp.Matrix([1, 0, 0]), sp.Matrix([0, 0, 1]), coords)
display(sp.simplify(K_13))

K_23 = DG503.sectional_curvature(GK_f1, sp.Matrix([[x1], [x2], [x3]]), sp.Matrix([0, 1, 0]), sp.Matrix([0, 0, 1]), coords)
display(sp.simplify(K_23))
```

```python
r, f, x1, x2, x3, x4 = sp.symbols('r f x1 x2 x3 x4', real=True)

f = sp.Function('f')(x1,x2)
coords = [x1, x2, x3, x4]

r = sp.Matrix([[x1], [x2], [x3], [x4], [f], [1]])
display(r)

jacobi_r = r.jacobian(coords)
display(jacobi_r)

G = jacobi_r.T * jacobi_r
display(G)
```

```python
det_G = G.det()
display(det_G)

Hes_G = sp.hessian(f, [x1,x2])

S = DG503.scalar_curvature(G, sp.Matrix([x1,x2,x3,x4]), coords)
display(sp.simplify(S))
```

```python
S_2 = 2*Hes_G.det()/det_G**2
display(sp.simplify(S_2))

display(sp.simplify(S - S_2))
```

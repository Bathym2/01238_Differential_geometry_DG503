# Differential Geometry (DG503) Library Examples
This notebook demonstrates the core functions implemented in `functions.py` for symbolic calculations in differential geometry.

```python
import sympy as sp
from functions import (
    calculate_jacobi_function, 
    calculate_signed_jacobi_function, 
    calculate_christoffel_symbols,
    calculate_lie_derivative_metric,
    calculate_g_dot
)

# Enable pretty printing for LaTeX output
sp.init_printing(use_latex='mathjax')
```

## 1. Jacobi Functions (Area Element)
Calculating the area element for a sphere of radius $R$ using spherical coordinates $(u, v)$ where $u=\phi, v=\theta$.

```python
R, u, v = sp.symbols('R u v', real=True, positive=True)
r_sphere = sp.Matrix([
    R * sp.sin(u) * sp.cos(v),
    R * sp.sin(u) * sp.sin(v),
    R * sp.cos(u)
])

jacobi = calculate_jacobi_function(r_sphere, [u, v])
print("Jacobi Function (Area Element):")
jacobi
```

## 2. Christoffel Symbols
Calculating the Christoffel symbols of the second kind $\Gamma^k_{ij}$ for the sphere metric.

```python
# Define the standard metric for a sphere
g_sphere = sp.Matrix([
    [R**2, 0],
    [0, R**2 * sp.sin(u)**2]
])

coords = [u, v]
chris = calculate_christoffel_symbols(g_sphere, coords)

print("Non-zero Christoffel Symbols:")
for indices, value in chris.items():
    if value != 0:
        i, j, k = indices
        display(sp.MatMul(sp.Symbol(f'\\Gamma^{k}_{{{i}{j}}}'), sp.Symbol('=')), value)
```

## 3. Lie Derivative
Calculating the Lie derivative of the metric $g$ with respect to a vector field $X$.

```python
# Define a vector field X (e.g., a rotation field)
X = sp.Matrix([0, 1]) 

lie_derivative = calculate_lie_derivative_metric(g_sphere, X, coords)
print("Lie Derivative of the metric w.r.t X:")
lie_derivative
```
